b. Outlier Treatment using the Z-score method

```python
data1
print("the mean of columns:\n")
mean=np.mean(data1)
print(mean)
```

```python
print("std of columns\n")
std=np.mean(data1)
print(std)
```

```python
a1= data1.quantile(0.30)
a2 = data1.quantile(0.70)
iqr = a2 - a1
outliers = data1[((data1<(a1-1.5*iqr)) | (data1>(a2+1.5*iqr)))]
print("number of outliers:" + str(len(outliers)))
print("min outlier:" + str(outliers.min()))
print("max outlier:" + str(outliers.max()))

from scipy import stats
z_scores = stats.zscore(outliers)
abs_z_scores = np.abs(z_scores)
filtered_entreis = (abs_z_scores < 3).all(axis=1)
print(filtered_entreis)
```

c. Deal with correlated variables

```python
data1.corr()
```

```python
def correlation(df, threshold):
    col_corr = set() 
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] 
                col_corr.add(colname)
print("corr",correlation)
print("correlation",threshold)
```

Q3 Perform dimensionality reduction using PCA such that the 95% of the variance is explained

```python
x = data1.drop(['TENURE'],axis = 1)
y = data1['TENURE']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
accuracy_score(y_test,prediction)
```

```python
confusion_matrix(y_test,prediction)
```

```python
result= classification_report(y_test,prediction)
print(result)
```

```python
#PCA components
from sklearn.decomposition import PCA
model=PCA()
```

```python
# fitting the model to train the data
model.fit(x_train)
```

```python
#Transforming the model for better result
xtrain_transformed=model.transform(x_train)
print(xtrain_transformed.shape)
```

```python

xtest_transformed=model.transform(x_test)
print(xtest_transformed.shape)
```

**Q4. Find the optimum value of k for k-means clustering using the elbow method. Plot the elbow curve**

```python

from sklearn.cluster import KMeans
```

```python
cluster=[]  

for i in range(1,7):
  kmeans=KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10, 
                max_iter=300, 
                tol=0.0001,  
                verbose=0, 
                random_state=None, 
                copy_x=True,  
                algorithm='auto')
  
  kmeans.fit(data1)

  cluster.append(kmeans.inertia_)  
```

```python
plt.plot(range(2,8),cluster)
```

```python
plt.plot(range(1,7),cluster)
```
